2. histogram of observationsmjd (weekly)

```python
df['observationStartMJD'] = pd.to_datetime(df['observationStartMJD'], unit='D', origin=pd.Timestamp('1858-11-17'))
```

```python
df.head()
```

```python
df['Year'] = df['observationStartMJD'].dt.year

# Group by year and plot histograms
for year, group in df.groupby('Year'):
    plt.figure(figsize=(8, 6))
    plt.hist(group['observationStartMJD'], bins=52)
    plt.title(f'ObservationStartMJD Distribution for Year {year}')
    plt.xlabel('ObservationStartMJD')
    plt.ylabel('Frequency')
    plt.grid(True)
    plt.show()
```

3. ra vs dec(equatorial coord system) -find equation- ecliptic coordinate system(lambda vs beta) -> plot lambda vs beta

```python
from astropy.coordinates import SkyCoord
import astropy.units as u
```

```python
# Convert equatorial coordinates to SkyCoord object
equatorial_coords = SkyCoord(ra=df['fieldRA'], dec=df['fieldDec'], unit='deg', frame='icrs')
```

```python
# Convert equatorial coordinates to ecliptic coordinates
ecliptic_coords = equatorial_coords.transform_to('barycentrictrueecliptic')
```

```python
# Extract lambda and beta values
lambda_values = ecliptic_coords.lon.deg
beta_values = ecliptic_coords.lat.deg

# Plot lambda vs beta
plt.figure(figsize=(8, 6))
plt.scatter(lambda_values, beta_values, color='blue',s=0.001)
plt.title('Ecliptic Coordinates (lambda vs beta)')
plt.xlabel('Lambda (degrees)')
plt.ylabel('Beta (degrees)')
plt.grid(True)
plt.show()
```
